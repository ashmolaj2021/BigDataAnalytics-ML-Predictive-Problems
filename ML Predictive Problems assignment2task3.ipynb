```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import fetch_20newsgroups_vectorized
data = fetch_20newsgroups_vectorized(subset='train')
X = data['data']
Y = data['target']
```

```python
from sklearn.cluster import KMeans, MiniBatchKMeans

kmeans = KMeans(n_clusters=20)
clusters = kmeans.fit_predict(X)

res = list(zip(*np.unique(clusters, return_counts=True)))
```

```python
res
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
cnf = confusion_matrix(Y, clusters)
sns.heatmap(cnf, yticklabels=data['target_names'],
            xticklabels=["cluster {}".format(i) for i in range(20)])
```
